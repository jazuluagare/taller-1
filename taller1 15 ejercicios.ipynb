```python
# Ejercicio 5

x=int(input('digite un numero: '))
y=[] #se crea una lista

x=abs(x) #se saca el valor absoluto de el numero ingresado

while x>=10:
    r=x%10 #sacamos el residuo
    y.append(r)
    x=x//10# sacamos el numero entero de la division
y.append(x)

z=[]#lista de residuos
for i in range (0,len(y)):
    z.append(y[-i-1])
c=0

d=[] #numero de 5 consecutivos
for n in range(0,len(z)):
    if z[n]==5:
        c+=1
        if n==len(z)-1:
            d.append(c)
        else:
            if c!=0:
               d.append(c)
               c=0 #reiniciamos el contador si no hay 5
print (d)
```

```python
# Ejercicio 6
    
def suma_digitos(x):# se define una funcion del residuo
    cont=0
    while x>0:
        residuo=x%10
        x=x//10
        if residuo>0:
           cont=cont+residuo
    return cont
def numero_fibonacci(x):# se define una nueva funcion
    num1=0
    num2=1
    resultado= False
    while(num1<=x):
        if (num1==x):
            resultado= True
        suma=num1+num2
        num1=num2
        num2=suma
    return resultado

while True:#ciclo cuando la condicion es verdadera permite ingresar un numero
    #entero y corrige si es negativo
    try:
        num=int(input('escriba un número entero = '))
        break
    except ValueError:
        print('el valor ingresado es incorrecto. ')
        
cont=suma_digitos(num)
if cont<0:
   cont=cont*(-1)
    
resultado=numero_fibonacci(cont)
if resultado:
    print('la suma de los digitos en valor absoluto de numero ingresado \
          es',cont,'.\nY el numero',cont,'es dela serie fibonacci.')
else:
    print('la suma de los digitos en valor absoluto es el numero',cont,'\n\
          Y no es de la serie fibonacci.')
```

```python
# Ejercicio 11

x=input('escriba una cadena de texto:')

a=0 #contador de vocales mayusculas
b=0 #contador de tildes
c=0 #contador de digitos
d=0 #contador de espacios
e=0 #contador de palabras reservadas

q=['and','del','for','is','raise','assert','if','else','elif','from','lambda',\
   'return','break','global','not','try','class','except','or','while',\
   'continue','exec','import','yield','def','finally','in','print']

for i in range(0,len(x)):
    if x[i].isupper()==True: #identificamos que sea vocal y mayuscula
        if ord(x[i])==65 or ord(x[i])==69 or ord(x[i])==73 or ord(x[i])==79\
        or  ord(x[i])==85 or ord(x[i])==193 or ord(x[i])==201 or \
        ord(x[i])==205 or ord(x[i])==211 or ord(x[i])==218:
            
            a+=1 # sumardor contador de vocales mayusculas
    if ord(x[i])==225 or ord(x[i])==233 or ord(x[i])==237 or ord(x[i])==243\
        or  ord(x[i])==250 or ord(x[i])==193 or ord(x[i])==201 or \
        ord(x[i])==205 or ord(x[i])==211 or ord(x[i])==218:  
        #miramos si es un caracter con tilde 
            b+=1#sumador de tildes
            
    if x[i].isdigit()==True: #contamos los digitos
        
        c+=1 #sumador de digitos
    if ord(x[i])==32:# contamos los espacios
        d+=1
        
k= x.split() #concertimos la cadena en una lista
for n in range(len(k)):#iteramos en la lista creada dependiendo de su \
    #longitud
    if k[n] in q:
        e+=1
        
print(f'en la cadena hay {a} vocales mayusculas')
print(f'en la cadena hay {b} tildes')       
print(f'en la cadena hay {c} digitos')        
print(f'en la cadena hay {d} espacios')
print(f'en la cadena hay {e} palabras reservadas')
        
```

```python
# Ejercicio 16

#digitar el numero de caracteres de la lista
long = int(input('digite la longitud de la lista: '))
lista = []
for i in range(long):   #ciclo para agregar cada termino de la lista
    lista.append(float(input('inserte un numero: ')))

lis_orden = []  #defino una nueva lista ordenada
lis_orden = sorted(lista)   #la lista sera la ingresada ordenada ascendente
          
if lis_orden == lista:  #condicional de que la lista ingresada sea ordenada
    print('la lista esta ordenada')

    #ingresar el numero que se desea añadir
    numero = float(input('inserte un numero para agregar a la lista: '))    
    lista.append(numero)    #añadirlo a la lista
    lis_orden = sorted(lista)   #organizar la nueva lista ascendentemente
    posicion = lis_orden.index(numero)  #dar la posicion del numero ingresado
    print('la posicion indexada del numero inresado es: ',posicion)

else:
    print('la lista no esta ordenada')
```
